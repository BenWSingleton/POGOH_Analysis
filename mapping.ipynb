```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point, LineString
import folium
```

```python
df = pd.read_parquet('./data/POGOH/combined data/data.parquet')
```

```python
df['Starting Point'] = df.apply(lambda row: Point(row['Start Station Longitude'], row['Start Station Latitude']), axis=1)
df['Ending Point'] = df.apply(lambda row: Point(row['End Station Longitude'], row['End Station Latitude']), axis=1)
```

```python
df['trip'] = df.apply(lambda row: LineString([row['Starting Point'], row['Ending Point']]), axis=1)
```

```python
trip_counts = df['trip'].value_counts().reset_index()
```

```python
gdf = gpd.GeoDataFrame(df, geometry='trip')
```

```python
gdf = gdf[(gdf['Start Date'].dt.year==2024) & ((gdf['Start Station Neighborhood']=='Strip District') | (gdf['End Station Neighborhood']=='Strip District'))]
```

```python
trips = folium.Map(location=[df['Start Station Latitude'].mean(), df['Start Station Longitude'].mean()], zoom_start=12)
```

```python
for _, row in gdf.iterrows():
    folium.PolyLine(locations=[(row['Start Station Latitude'], row['Start Station Longitude']), (row['End Station Latitude'], row['End Station Longitude'])], color='blue').add_to(trips)
```
